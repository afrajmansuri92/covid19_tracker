```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/sampel2.jpg')
```

```python
import face_recognition as fr
```

```python
locs = fr.face_locations(image)
```

```python
print(len(locs), " faces detected")
```

```python
for face in locs:
  cv2.rectangle(image, (face[3], face[0]), (face[1], face[2]), (0, 255, 0), 2)
```
